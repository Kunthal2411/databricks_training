```python
Task for today;

1. create a df from samples.tpch.customer
2. rename all col name. (without c_)
3. add new col called ingestion_date
4. save the new df to table in your schema (dev.default)
```

```python
df=spark.read.table("samples.tpch.customer")
```

```python
df.display()
```

```python
df_renamed=df.withColumnsRenamed({"c_custkey":"custkey","c_name":"name","c_address":"address","c_nationkey":"nationkey","c_phone":"phone","c_acctbal":"acctbal","c_mktsegment":"mktsegment","c_comment":"comment"})
df_renamed.display()
```

```python
from pyspark.sql.functions import *
```

```python

 df_new_col=df_renamed.withColumn("Ingestion Date",current_timestamp())


```

```python
df_new_col.display()
```

```python
df_dinal=df_new_col.withColumnRenamed("Ingestion Date","ingestion_date")
```

```python
df_dinal.display()
```

```python
df_dinal.write.saveAsTable("kj_dev.default.kj_customer")
```
