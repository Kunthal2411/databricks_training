```sql
select * from samples.tpcds_sf1.call_center LIMIT 10

select distinct cc_call_center_sk,cc_call_center_id from samples.tpcds_sf1.call_center

select cd_education_status,cd_gender,cd_marital_status,count(*) from samples.tpcds_sf1.customer_demographics
group by  cd_education_status,cd_gender,cd_marital_status
having count(*) >1

select cc.cc_call_center_sk, cc.cc_call_center_id, r.r_reason_id, r.r_reason_desc
from samples.tpcds_sf1.call_center cc
join samples.tpcds_sf1.reason r
  on cc.cc_call_center_sk = r.r_reason_sk
limit 10

```
